```python
import matplotlib.pyplot as plt

# Data: (Money Supply Growth %, Inflation Rate %, Decade Label)
data = [
    (3.0, 0.0, '1870s'),
    (6.0, 0.0, '1880s'),
    (4.0, 0.0, '1890s'),
    (6.0, 1.0, '1900s'),
    (8.0, 6.0, '1910s'),
    (4.0, 0.0, '1920s'),
    (2.0, -1.0, '1930s'),
    (10.0, 6.0, '1940s'),
    (6.0, 2.0, '1950s'),
    (7.0, 3.0, '1960s'),
    (11.0, 7.0, '1970s'),
    (8.0, 4.0, '1980s'),
    (5.0, 2.0, '1990s'),
    (6.0, 2.0, '2000s')
]

# Unpack data
x = [d[0] for d in data]  # Money supply growth
y = [d[1] for d in data]  # Inflation rate
labels = [d[2] for d in data]

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='green', marker='s')

# Add labels to each point
for i in range(len(data)):
    plt.text(x[i] + 0.2, y[i], labels[i], fontsize=9)

# Axis labels and title
plt.xlabel('Growth in money supply (percent)')
plt.ylabel('Inflation rate (percent)')
plt.title('Money Supply Growth vs Inflation Rate by Decade')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)

# Show plot
plt.tight_layout()
plt.show()
```
