```python
# Monopoly profit-maximization with YOUR numbers
# a=10, b=1, c=2  â†’  Y* = 4,  P* = 6

import numpy as np
import matplotlib.pyplot as plt

# -------------------------------------------------
# PARAMETERS (YOUR EXAMPLE)
a = 10          # Demand intercept
b = 1           # Demand slope
c = 2           # Marginal cost (constant)

# -------------------------------------------------
# QUANTITY RANGE
Y = np.linspace(0, 12, 500)

# CURVES
P  = a - b * Y                # Demand: P(Y) = 10 â€“ Y
MR = a - 2*b * Y              # MR = 10 â€“ 2Y
MC = np.ones_like(Y) * c      # MC = 2

# OPTIMAL POINT
Y_star = (a - c) / (2 * b)    # Y* = (10-2)/2 = 4
P_star = (a + c) / 2          # P* = (10+2)/2 = 6

# ELASTICITY & MARKUP (for annotation)
eta_abs = P_star / (b * Y_star)          # |Î·| = P/(bY)
markup  = (P_star - c) / P_star          # (P-MC)/P

# -------------------------------------------------
# PLOT
plt.figure(figsize=(10, 6))

# Demand
plt.plot(Y, P,  label='Demand: $P(Y)=10-Y$', color='blue', linewidth=2.5)

# MR
plt.plot(Y, MR, label='Marginal Revenue: $MR=10-2Y$', color='red',
         linestyle='--', linewidth=2.5)

# MC
plt.hlines(c, 0, 12, color='green', linewidth=2.5,
           label='Marginal Cost: $MC=2$')

# Optimal point
plt.scatter([Y_star], [P_star], color='black', s=120, zorder=5)
plt.vlines(Y_star, 0, P_star, color='black', linestyle=':', alpha=0.7)
plt.hlines(P_star, 0, Y_star, color='black', linestyle=':', alpha=0.7)

# Annotation box
plt.text(Y_star + 0.5, P_star + 0.5,
         f'$Y^* = {Y_star}$\n$P^* = {P_star}$\n'
         f'$|\\eta| = {eta_abs:.1f}$\nMarkup = {markup:.2f}',
         fontsize=11, bbox=dict(facecolor='wheat', edgecolor='black', boxstyle='round,pad=0.5'))

plt.text(5.5, 3, 'MR = MC', fontsize=11, color='red', fontweight='bold')

# Labels & title
plt.title('Monopoly Example (a=10, b=1, c=2)\n'
          '$Y^* = \\frac{a-c}{2b}=4$,  $P^* = \\frac{a+c}{2}=6$',
          fontsize=14, fontweight='bold', pad=20)
plt.xlabel('Quantity (Y)', fontsize=12)
plt.ylabel('Price (P)', fontsize=12)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.xlim(0, 12)
plt.ylim(0, 12)

plt.tight_layout()
# print(f"Y* = {Y_star}, P* = {P_star}, |Î·| = {eta_abs:.1f}, Markup = {markup:.2f}")
plt.show()
```

# Permanent Income Hypothesis: 

```python
Math(r"C_t = \frac{1}{T+1} \sum_{j=0}^{T} Y_{t+j}")
```

```python
Math(r"C_t = \frac{1}{2} (Y_t + Y_{t+1})")
```

```python
# Visualizes PIH / Life-Cycle example from Garin/Lester/Sims (2021)
# T=50, transitory vs permanent income shock

import numpy as np
import matplotlib.pyplot as plt

# === PARAMETERS ===
T = 50                  # Lifetime periods
shock_year = 25         # Year of income shock
baseline_income = 1.0   # Normal income
transitory_bonus = 1.0  # +1 in one year
permanent_increase = 1.0  # +1 forever after shock_year

# === 1. BASELINE: Y = 1, C = 1 ===
years = np.arange(1, T + 1)
Y_baseline = np.ones(T) * baseline_income
C_baseline = np.ones(T) * baseline_income

# === 2. TRANSITORY SHOCK: Y â†‘ 1 in year 25, then back ===
Y_transitory = Y_baseline.copy()
Y_transitory[shock_year - 1] += transitory_bonus  # +1 in year 25

# Average lifetime income after shock
remaining_years = T - shock_year + 1
total_income_trans = np.sum(Y_transitory)
avg_income_trans = total_income_trans / T
C_transitory = np.ones(T) * avg_income_trans  # C = average lifetime income

# === 3. PERMANENT SHOCK: Y â†‘ 1 from year 25 onward ===
Y_permanent = Y_baseline.copy()
Y_permanent[shock_year - 1:] += permanent_increase

total_income_perm = np.sum(Y_permanent)
avg_income_perm = total_income_perm / T
C_permanent = np.ones(T) * avg_income_perm  # C = 2

# === 4. PLOT ===
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 9), sharex=True)

# --- TRANSITORY SHOCK ---
ax1.plot(years, Y_transitory, label='Income (Y) - Transitory Shock', color='blue', linewidth=2)
ax1.plot(years, C_transitory, label='Consumption (C)', color='red', linestyle='--', linewidth=2.5)
ax1.axvspan(shock_year, shock_year, color='gray', alpha=0.3, label='Shock Year')
ax1.set_title('Transitory Income Shock (Bonus in Year 25)\nMPC = 1/50 = 0.02', fontsize=14, fontweight='bold')
ax1.set_ylabel('Income / Consumption')
ax1.legend()
ax1.grid(True, alpha=0.3)
ax1.annotate('Y â†‘ 1 â†’ C â†‘ 0.02\n(MPC = 0.02)', xy=(shock_year, 1.5), xytext=(30, 1.6),
             arrowprops=dict(arrowstyle='->', color='black'), fontsize=11, color='blue')

# --- PERMANENT SHOCK ---
ax2.plot(years, Y_permanent, label='Income (Y) - Permanent Shock', color='blue', linewidth=2)
ax2.plot(years, C_permanent, label='Consumption (C)', color='red', linestyle='--', linewidth=2.5)
ax2.axvspan(shock_year, T, color='gray', alpha=0.3, label='Shock Period')
ax2.set_title('Permanent Income Shock (Income â†‘1 from Year 25)\nMPC = 1.0', fontsize=14, fontweight='bold')
ax2.set_xlabel('Year')
ax2.set_ylabel('Income / Consumption')
ax2.legend()
ax2.grid(True, alpha=0.3)
ax2.annotate('Y â†‘ 1 â†’ C â†‘ 1\n(MPC = 1.0)', xy=(shock_year, 1.5), xytext=(35, 1.8),
             arrowprops=dict(arrowstyle='->', color='black'), fontsize=11, color='blue')

# === 5. FINAL TOUCHES ===
plt.suptitle('Permanent Income Hypothesis: Life-Cycle Smoothing (T=50)\n'
             'Garin, Lester, Sims (2021) - Two-Period Model Generalized', 
             fontsize=16, fontweight='bold', y=0.98)
plt.tight_layout(rect=[0, 0, 1, 0.95])

plt.show()
```

```python
# Fisher Effect: Nominal Fed Funds vs Inflation (1960â€“2016)
# Correlation = 0.70 | Garin/Lester/Sims (2021), Ch. 21

import pandas as pd
import matplotlib.pyplot as plt
from pandas_datareader import data as pdr
import warnings
warnings.filterwarnings("ignore")

print("Fetching FRED data: FEDFUNDS, CPIAUCSL...")

# === 1. DATA ===
start = '1960-01-01'
end = '2016-12-31'

# Nominal: Federal Funds Rate (monthly average)
fed_funds = pdr.get_data_fred('FEDFUNDS', start, end)

# Inflation: CPI YoY % change
cpi = pdr.get_data_fred('CPIAUCSL', start, end)
cpi_yoy = cpi.pct_change(12) * 100  # YoY inflation

# Merge
df = pd.concat([fed_funds, cpi_yoy], axis=1).dropna()
df.columns = ['Nominal_Rate', 'Inflation']

# === 2. CORRELATION ===
corr = df['Nominal_Rate'].corr(df['Inflation'])
print(f"Correlation (1960â€“2016): {corr:.3f}")

# === 3. PLOT ===
plt.figure(figsize=(11, 7))
ax = plt.gca()

# Scatter
ax.scatter(df['Inflation'], df['Nominal_Rate'], 
           c='steelblue', s=60, alpha=0.7, edgecolors='black', linewidth=0.5, label='Monthly Data')

# 45Â° line (Fisher: i = Ï€^e if r constant)
min_val = df.min().min() - 1
max_val = df.max().max() + 1
ax.plot([min_val, max_val], [min_val, max_val], 'r--', linewidth=2, label='45Â° Line (r = 0)')

# Regression line
z = np.polyfit(df['Inflation'], df['Nominal_Rate'], 1)
p = np.poly1d(z)
ax.plot(df['Inflation'], p(df['Inflation']), "k-", alpha=0.8, linewidth=2, 
        label=f'Fit: i = {z[0]:.2f}Ï€ + {z[1]:.2f}')

# Labels
ax.set_xlabel('Inflation Rate (Ï€, % YoY)', fontsize=13)
ax.set_ylabel('Nominal Interest Rate (i, %)', fontsize=13)
ax.set_title('Fisher Effect: U.S. Federal Funds Rate vs Inflation (1960â€“2016)\n'
             'Correlation = 0.70 | Garin, Lester, Sims (2021)', 
             fontsize=14, fontweight='bold')

ax.grid(True, alpha=0.3)
ax.legend(fontsize=11)
ax.set_xlim(min_val, max_val)
ax.set_ylim(min_val, max_val)

plt.tight_layout()
plt.show()
```
